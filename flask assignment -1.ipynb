In Flask, app routing refers to the process of defining URL routes and associating them with specific functions or views that handle the corresponding HTTP requests. Routes are used to map URLs to the functions that generate the content to be displayed on the web application.

In Flask, app routing is typically done using the @app.route() decorator. The decorator specifies the URL path (route) and the HTTP methods (e.g., GET, POST, etc.) that the function should handle. When a user makes a request to a specific URL on the Flask web application, the routing mechanism determines which function should be executed based on the URL path and the HTTP method of the request.




-Why do we use app routes?

App routes in Flask serve several important purposes:

1.Handling Different URLs: Routes allow you to handle different URLs within your web application and provide a way to execute different functions or views based on the URL visited by the user.

2.Structuring the Application: Using app routes helps in organizing your Flask application by separating different functionalities into their respective views, making the code more maintainable and understandable.

3.RESTful APIs: For web applications that serve as RESTful APIs, routes are essential for defining endpoints and associating them with specific functions to handle different API operations (e.g., GET, POST, PUT, DELETE).

4.Request Methods: Routes also enable handling different HTTP methods (e.g., GET, POST, etc.) on the same URL, allowing you to implement different actions based on the type of request.

5.URL Generation: Flask provides a feature called "url_for()" that generates URLs based on the route functions, making it easier to manage links and navigate within the application.

In Flask, the url_for() function is used for URL building. It generates a URL for a given route function based on its name and any arguments passed to it. This function is useful when you want to create URLs dynamically, especially when you change the URL structure or route names in your Flask application.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'This is the home page.'

@app.route('/contact')
def contact():
    return 'This is the contact page.'

@app.route('/links')
def links():
    home_url = url_for('home')     
    contact_url = url_for('contact')  

    return f"Home url : {home_url}\nContact url : {contact_url}"

if __name__ == '__main__':
    app.run()
```
